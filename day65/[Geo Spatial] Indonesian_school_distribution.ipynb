## 1. Import modules

```python
import matplotlib.pyplot as plt
import geopandas as gpd
import pandas as pd
import numpy as np
```

## 2. Data Preparation

```python
# Import school participation
path = './Angka Partisipasi Sekolah.xlsx'
df = pd.read_excel(path)
display(df.head())
```

```python
# Import GeoJSON Data
path = './gadm36_IDN_1.json'
df_geo = gpd.read_file(path)
display(df_geo.head())
```

gpd.read_file will directly create a GeoDataFrame class

```python
# Check kind of object districts is
print('df_geo data type: {}'.format(type(df_geo)))

# Plot GeoDataFrame
print("\nPlotting GeoDataFrame: ")
df_geo.plot()
```

## 3. Merge Datasets

The geometry from df_geo will appended into df for displaying the distribution

```python
df_join = df_geo.merge(df, how='inner', left_on='NAME_1', right_on='Province')
df_join
```

```python
df_join = df_join[['Province', '7-12', '13-15', '16-18', '19-24', 'geometry']]
df_join.head()
```

## 4. Create Map

Creating choropleth map

```python
# Choose what we want to visualize
values = '16-18'

# Set the value range for the choropleth
vmin, vmax = df_join[values].min(), df_join[values].max()

# Create figure and axes for matplotlib
fig, ax = plt.subplots(1, figsize=(50, 70))

# Create map
df_join.plot(column=values, cmap='Oranges', linewidth=0.8, ax=ax, edgecolor='0.8')
```

The shaded color is telling the number of student between 16-18 y.o. in Indonesia

### A. Customize Map

Adding the details of map (title, color bar, annotation)

```python
# Set variabel we want to visialize
values = '16-18'

# Set the value range for the choropleth
vmin, vmax = 0, 100

# Create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(30,20))

# Remove the axis
ax.axis('off')

# Add title
title = 'School Participations Rate Based on Age {}'.format(values)
ax.set_title(title, fontdict={'fontsize':'25', 'fontweight':'3'})

# Create annotation for the data source
ax.annotate('Source: Badan Pusat Statistik Indonesia', xy=(0.1, .08), xycoords='figure fraction', 
            horizontalalignment='left', verticalalignment='top', fontsize=12, color='#555')

# Create colorbar as a legend
sm = plt.cm.ScalarMappable(cmap='Oranges', norm=plt.Normalize(vmin=vmin, vmax=vmax))

# Add colorbar to the figure
cbar = fig.colorbar(sm)

# Create map
df_join.plot(column=values, cmap='Oranges', linewidth=0.8, ax=ax, edgecolor='0.8', 
             norm=plt.Normalize(vmin=vmin, vmax=vmax))
```

## 5. Generate All Map

```python
# Ignoring the warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
def build_map(df_geo, values):
    # Set variabel we want to visialize
    values = df_geo.columns[values]

    # Set the value range for the choropleth
    vmin, vmax = 0, 100

    # Create figure and axes for Matplotlib
    fig, ax = plt.subplots(1, figsize=(40,20))

    # Remove the axis
    ax.axis('off')

    # Add title
    title = 'School Participations Rate Based on Age {}'.format(values)
    ax.set_title(title, fontdict={'fontsize':'45', 'fontweight':'10'})

    # Create annotation for the data source
    ax.annotate('Source: Badan Pusat Statistik Indonesia', xy=(0.1, .08), xycoords='figure fraction', 
                horizontalalignment='left', verticalalignment='top', fontsize=30, color='#555')

    # Create colorbar as a legend
    sm = plt.cm.ScalarMappable(cmap='Oranges', norm=plt.Normalize(vmin=vmin, vmax=vmax))

    # Add colorbar to the figure
    cbar = fig.colorbar(sm)

    # Create map
    df_geo.plot(column=values, cmap='Oranges', linewidth=0.8, ax=ax, edgecolor='0.8',
                norm=plt.Normalize(vmin=vmin, vmax=vmax))

    # Add labels
    df_geo['coords'] = df_geo['geometry'].apply(lambda x:x.representative_point().coords[:])
    df_geo['coords'] = [coords[0] for coords in df_geo['coords']]
    for idx, row in df_join.iterrows():
        plt.annotate(text=row['Province'], xy=row['coords'], horizontalalignment='center', color='Black', fontsize=18)
```

```python
for i in range(1, 5):
    build_map(df_join, i)
```
