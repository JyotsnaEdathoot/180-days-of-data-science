## Import modules

```python
from IPython.display import  display
import pandas as pd
import numpy as np 
url = "https://raw.githubusercontent.com/rromanss23/Machine_Leaning_Engineer_Udacity_NanoDegree/master/projects/customer_segments/customers.csv"

data = pd.read_csv(url)
data.drop(['Region', 'Channel'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data.shape
```

```python
display(data.describe())
```

```python
new_data = data.drop('Grocery', axis=1)

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(new_data, data.Grocery, 
                                                    test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor()
regressor = regressor.fit(X_train, y_train)
prediction = regressor.predict(X_test)
```

```python
prediction
```

```python
from sklearn.metrics import  r2_score
score = r2_score(y_test, prediction)
print("Predictions score: {}".format(score))
```
