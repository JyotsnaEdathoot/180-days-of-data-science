## Predictions

This is how to use the model to predict picture, for the example I took dog

```python
import matplotlib.pyplot as plt
import numpy as np
import keras
import cv2
```

```python
categories = ['Cat', 'Dog']

def image(path):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    plt.imshow(img)
    new_arr = cv2.resize(img, (60, 60))
    new_arr = np.array(new_arr)
    new_arr = new_arr.reshape(-1, 60, 60, 1)
    return new_arr

model = keras.models.load_model('catvsdog-60x60.model')
```

```python
predictions = model.predict([image('test/gbr.jpg')])
print(categories[predictions.argmax()])
```

```python
predictions2 = model.predict([image('test/gbr2.jpg')])
print(categories[predictions2.argmax()])
```

```python
predictions3 = model.predict([image('test/gbr3.jpg')])
print(categories[predictions3.argmax()])
```
