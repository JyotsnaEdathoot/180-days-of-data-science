## Importing Modules and Datasets

```python
import pandas as pd
```

```python
logins = pd.read_csv('datasets/logins.csv')
logins.info()
```

```python
logins.head()
```

## Rule 1: Must be at least 10 character length

```python
length_pass = logins['password'].str.len() >= 10
length_pass.head()
```

```python
valid_passwd = logins[length_pass]
valid_passwd.head()
```

```python
bad_passwd = logins[~length_pass]
bad_passwd.head()
```

```python
bad_passwd.shape[0], valid_passwd.shape[0]
```

## Rule 2: one lower case, one upper case, one numeric, one non-alphanumeric

```python
low_case = valid_passwd['password'].str.contains('[a-z]')
up_case = valid_passwd['password'].str.contains('[A-Z]')
num_case = valid_passwd['password'].str.contains('\d')
nonum_case = valid_passwd['password'].str.contains('\W')
```

```python
pd.concat([valid_passwd, low_case, up_case, num_case, nonum_case], axis=1)
```

```python
char_check = low_case & up_case & num_case & nonum_case

bad_passwd = bad_passwd.append(valid_passwd[~char_check], ignore_index=True)
valid_passwd = valid_passwd[char_check]
```

```python
bad_passwd.shape[0], valid_passwd.shape[0]
```

## Rule 3: Must not contains the phrase password (case insensitve)

```python
banned_phrases = valid_passwd['password'].str.contains('password', case=False)

bad_passwd = bad_passwd.append(valid_passwd[banned_phrases], ignore_index=True)
valid_passwd = valid_passwd[~banned_phrases]
```

```python
bad_passwd.shape[0], valid_passwd.shape[0]
```

## Rule 4: must not contain the user's first or last name

```python
valid_passwd['first_name'] = valid_passwd['username'].str.extract('(^[a-z]+)', expand = False)
valid_passwd['last_name'] = valid_passwd['username'].str.extract('([a-z]+$)')
```

```python
valid_passwd.head()
```

```python
for i, row in valid_passwd.iterrows():
    if row.first_name in row.password.lower() or row.last_name in row.password.lower():
        print(row)
        valid_passwd = valid_passwd.drop(index=i)
        bad_passwd = bad_passwd.append(row, ignore_index=True)
```
