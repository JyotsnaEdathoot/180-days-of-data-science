## BERT Extractive Summarizer

Summarize the text using BERT extractive summarizer for getting important and usefule sentences from the complete text

Story source: [link](https://twinfinite.net/2018/10/assassins-creed-odyssey-story-summary/5/)

```python
from summarizer import Summarizer

f = open('odyssey.txt', 'r')
full_text = f.read()

model = Summarizer()
result = model(full_text, min_length=60, max_length=500, ratio=0.4)

summarized_text = ''.join(result)
print(summarized_text)
```

## Keyword Extraction

```python
from nltk.corpus import stopwords
import itertools
import string
import pprint
import pke
import re

def get_nouns_multipartite(text):
  out = []

  extractor = pke.unsupervised.MultipartiteRank()
  extractor.load_document(input=text)

  pos = {'PROPN'} # VERB, ADJ, NOUN
  
  stoplist = list(string.punctuation)
  stoplist += ['-lrb-', '-rrb-', '-lcb-', '-rcb-', '-lsb-', '-rsb-']
  stoplist += stopwords.words('english')
  extractor.candidate_selection(pos=pos, stoplist=stoplist)

  extractor.candidate_weighting(alpha=1.1,
                                threshold=0.75,
                                method='average') 
  keyphrases = extractor.get_n_best(n=20)

  for key in keyphrases:
    out.append(key[0])

  return out

keywords = get_nouns_multipartite(full_text)
print(keywords)

filtered_keys = []
for keyword in keywords:
  if keyword.lower() in summarized_text.lower():
    filtered_keys.append(keyword)

print('\n')
print(filtered_keys)
```
