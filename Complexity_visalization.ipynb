```python
import matplotlib.pyplot as plt
import numpy as np

def constant_time(n):
    return 1

def linear_time(n):
    return n

def logarithmic_time(n):
    return np.log2(n)

def linearithmic_time(n):
    return n * np.log2(n)

def quadratic_time(n):
    return n**2

def cubic_time(n):
    return n**3

def exponential_time(n):
    return 2**n

def factorial_time(n):
    return np.math.factorial(n)

# Range of input values
n_values = np.arange(1, 21)

# Plotting
plt.figure(figsize=(12, 8))

plt.plot(n_values, [constant_time(n) for n in n_values], label='Constant Time')
plt.plot(n_values, [linear_time(n) for n in n_values], label='Linear Time')
plt.plot(n_values, [logarithmic_time(n) for n in n_values], label='Logarithmic Time')
plt.plot(n_values, [linearithmic_time(n) for n in n_values], label='Linearithmic Time')
plt.plot(n_values, [quadratic_time(n) for n in n_values], label='Quadratic Time')
plt.plot(n_values, [cubic_time(n) for n in n_values], label='Cubic Time')
plt.plot(n_values, [exponential_time(n) for n in n_values], label='Exponential Time')
plt.plot(n_values, [factorial_time(n) for n in n_values], label='Factorial Time')

plt.xlabel('Input Size (n)')
plt.ylabel('Operations')
plt.title('Time Complexity Visualizations')
plt.legend()
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def f_big_o(n):
    return 2 * n**2 + 3 * n + 1

def f_omega(n):
    return 0.5 * n**2

def f_theta(n):
    return n**2 + 2 * n

# Range of input values
n_values = np.arange(1, 21)

# Plotting
plt.figure(figsize=(12, 8))

# Function for Big O notation
plt.plot(n_values, [f_big_o(n) for n in n_values], label='O(n^2)', linestyle='--')

# Function for Omega notation
plt.plot(n_values, [f_omega(n) for n in n_values], label='(n^2)', linestyle='-.')

# Function for Theta notation
plt.plot(n_values, [f_theta(n) for n in n_values], label='(n^2)')

plt.xlabel('Input Size (n)')
plt.ylabel('Operations')
plt.title('Big O, Omega, and Theta Notations')
plt.legend()
plt.grid(True)
plt.show()
```
