# Adding Text in Matplotlib

Tutorial: [Kimberly Fessel](https://www.youtube.com/watch?v=NBYzSaTbodM)

## Import Modules

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 1, 4, 6, 4])
```

## Add Text

### Basic Text

```python
plt.scatter(x, y)
plt.text(4, 5.5, "Peak Point", ha="center")
plt.text(2.2, 1, "Low Point", va="center")
```

## Styling

### Color and Font Properties

```python
students = ["A", "B", "C", "D", "E"]
grades = [75, 52, 58, 81, 67]

plt.bar(students, grades, edgecolor='black', color='xkcd:light gray blue')
plt.axhline(100, color="xkcd:brick red", lw=3)

plt.text(2.5, 104, "Maximum Grade",
         fontsize=14,
         color="xkcd:brick red",
         fontweight="bold",
         fontstyle="italic")

plt.xlabel("Students")
sns.despine()
```

## Bounding Box: `bbox`

```python
from matplotlib.patches import Circle

line_x = np.linspace(0, 10, 1001)
line_y = 5 - 2 * np.abs(line_x - 5)
```

```python
plt.figure(figsize=(6, 6))
plt.plot(line_x, line_y, lw=3)
plt.gca().margins(0.1)

plt.text(6, 5, "Critical Point",
         bbox={'facecolor':'red',
               'alpha':0.3})

cir = Circle((5,5), 0.5, color='red', lw=3, fill=False, zorder=10)
plt.gca().add_patch(cir)
```

## Text in General Position or at Variable

### General Postiion: `transform=ax.transAxes`

```python
def add_mean_line(data):
    mu = np.mean(data)
    
    ax = plt.gca()
    ax.axvline(mu, color='black', lw=3, ls='--')
    
    ax.text(0.7, 0.9, f"mean={mu:.2f}",
            transform=ax.transAxes, # Transform positions to range from (0,0) to (1,1)
            fontweight='bold', fontsize=12)
```

```python
values = np.random.normal(20, 2, size=100)

plt.hist(values, color='lightblue')
add_mean_line(values)
```

## Variable in a Loop

```python
students = ['A', 'B', 'C', 'D', 'E']
grades = [90, 75, 65, 88, 78]

plt.bar(students, grades, edgecolor='black', color='xkcd:light gray blue')

for idx, grade in enumerate(grades):
    plt.text(idx, grade-8, f"{grade}",
             fontsize=16, 
             ha='center') # idx=x, grade=y

plt.xlabel("Students")
plt.yticks([])
sns.despine(left=True)
```
