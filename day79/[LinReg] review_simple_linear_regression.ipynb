## Import the modules

```python
from sklearn.linear_model import LinearRegression
import random
```

## Create two empty lists

```python
feature_set = []
target_set = []
```

## Declare global variables

```python
# Get number of rows for the dataset
num_rows = 5000

# Limit the possible values
random_limit = 100
```

## Create the data set

```python
# Create the feature dataset
for i in range(0, num_rows):
    x = random.randint(0, random_limit)
    y = random.randint(0, random_limit)
    z = random.randint(0, random_limit)
    
    # Create linear function for the target data
    function = (10*x) + (2*y) + (3*z)
    
    # Append the data to the lists
    feature_set.append([x,y,z])
    target_set.append(function)
```

```python
feature_set
```

```python
target_set
```

## Create Linear Regression model

```python
model = LinearRegression()
model.fit(feature_set, target_set)
```

## Create the test data

```python
# function = (10*x) + (2*y) + (3*z)

test_set = [[4,5,0]] # Expected output = function(4,5,0) = (10*4) + (2*5) + (3*0) = 40 + 10 = 50
predictions = model.predict(test_set)
```

```python
print("Prediction:", predictions, ' Coefficients:', model.coef_)
```

## Model score

```python
model.score(feature_set, target_set)
```
