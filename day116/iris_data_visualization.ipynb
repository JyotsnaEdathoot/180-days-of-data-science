```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
sns.set(style='whitegrid', color_codes=True)
```

```python
url = "https://raw.githubusercontent.com/siddharthjain1611/Iris_dataset/master/Iris.csv"
iris = pd.read_csv(url)
```

```python
iris.head()
```

```python
iris['Species'].value_counts()
```

```python
iris.plot(kind='scatter', x='SepalLengthCm', y='SepalWidthCm')
```

```python
# Seaborn jointplot show bivariate scatterplots and unvariative histograms in the same figure
sns.jointplot(x='SepalLengthCm', y='SepalWidthCm', data=iris, size=5)
```

```python
iris.shape
```

```python
iris.info()
```

```python
# Seaborn FacetGrid help to colorize scatterplot by species
sns.FacetGrid(iris, hue='Species', size=5)\
    .map(plt.scatter, 'SepalLengthCm', 'SepalWidthCm')\
    .add_legend()
```

```python
# Individual feature in seaborn (boxplot)
sns.boxplot(x='Species', y='PetalLengthCm', data=iris)
```

```python
ax = sns.boxplot(data=iris, x='Species', y='PetalLengthCm')
ax = sns.stripplot(data=iris, x='Species', y='PetalLengthCm', jitter=True, edgecolor='blue')
```

```python
# Violin plot combines the benefits of previous two plots
sns.violinplot(x='Species', y='PetalLengthCm', data=iris, size=6)
```

```python
sns.FacetGrid(iris,hue='Species', size=8)\
    .map(sns.kdeplot, 'PetalLengthCm')\
    .add_legend()
```

```python
sns.pairplot(iris.drop('Id', axis=1), hue='Species', size=5)
```

```python
# Feature by species

iris.drop('Id', axis=1).boxplot(by='Species', figsize=(12,6))
```
