### 2.A. Displaying WordCloud

```python
from wordcloud import WordCloud, STOPWORDS
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```python
def cleaning_text(text, lemmatizer = WordNetLemmatizer(),
                  stop_words = stopwords.words('english')):
    my_stop_list = ['say', 'said', 'u', 'first', 'one', 'nan']
    stop_words.extend(my_stop_list)
    words = word_tokenize(text)
    words_filtered = []
    
    for word in words:
        if word not in stop_words and word.isalpha():
            words_filtered.append(lemmatizer.lemmatize(word))
    
    return words_filtered
```

## Visualize the article Short Description

```python
df.description
```

```python
word_descriptions = cleaning_text(''.join(str(df['description'].tolist())))

# Displaying ten of df_description after cleaned into list
(pd.Series(nltk.ngrams(word_descriptions, 5)).value_counts())[:10]
```

```python
wordcloud = WordCloud(background_color='white',
                      max_words=3000,
                      colormap='rainbow')

wordcloud.generate(','.join(word_descriptions))
plt.figure(figsize=(16, 16))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
